```python
import pandas as pd
import numpy as np
```

```python
tree = pd.read_csv("https://data.cityofnewyork.us/api/views/uvpi-gqnh/rows.csv?accessType=DOWNLOAD")
tree.head()
```

```python
from matplotlib import pyplot
import random

good = tree[tree['health'] == 'Good']
fair = tree[tree['health'] == 'Fair']
poor = tree[tree['health'] == 'Poor']

bins = np.linspace(0, 50, 50)

pyplot.figure(figsize=(15,8))
pyplot.hist(good['tree_dbh'], bins, alpha=0.5, color="green", label='Good')
pyplot.hist(fair['tree_dbh'], bins, alpha=0.7, color="orange", label='Fair')
pyplot.hist(poor['tree_dbh'], bins, alpha=0.8, color="tomato", label='Poor')

pyplot.xlabel("Tree Diameter (Inches)",fontsize=15)
pyplot.ylabel("Number of Trees",fontsize=15)
pyplot.title("NYC Tree Health and Diameter, 2015",fontsize=20)
legend = pyplot.legend(loc='upper right',fontsize=14,frameon=False,title="Tree Health")
legend.get_title().set_fontsize('16')
pyplot.show()
```

Figure 1. The distribution of tree diameters categorized by tree health. All three types of health seem to follow the same general trend in distributions of diameters.
