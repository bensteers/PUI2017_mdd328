```python
# List of List Lengths for each Distribution

normListN = []
poiListN = []
biListN = []
chiListN = []
expListN = []

for i in range(0,100):
    normListN.append(len(normDist[i]))
    poiListN.append(len(poiDist[i]))
    biListN.append(len(biDist[i]))
    chiListN.append(len(chiDist[i]))
    expListN.append(len(expDist[i]))

# List of Means for each Distribution

normListMean = []
poiListMean = []
biListMean = []
chiListMean = []
expListMean = []

for i in range(0,100):
    normListMean.append(normDist[i].mean())
    poiListMean.append(poiDist[i].mean())
    biListMean.append(biDist[i].mean())
    chiListMean.append(chiDist[i].mean())
    expListMean.append(expDist[i].mean())
```

```python
matplotlib.pyplot.scatter(normListN,normListMean);
plt.title("Means Against Sample Sizes for Normal Distributions");
plt.xlabel("Sample Size")
plt.ylabel("Mean")
```

## Figure 1.

Calculating the means of the different sized normal distribution samples exhibited the central limit theory as the means tended symmetrically towards 500. The law of large numbers is also exhibited as the larger the sample size, the closer the means get to their most likely outcome of 500.
